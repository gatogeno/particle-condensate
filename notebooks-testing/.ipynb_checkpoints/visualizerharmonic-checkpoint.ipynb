# Check harmonic potential

```python
#Load the modules needed
import numpy as np
import matplotlib.pyplot as plt
import scipy
from colorsys import hls_to_rgb
from matplotlib.colors import ListedColormap, LogNorm
from matplotlib import cm
import seaborn as sns

import pylab as pyl

import sys
sys.path.append('../condpartsimulator/')
from utilities import storer, grid, fields
```

```python
#Load parameters used and grid
parameters=storer.reader("outputfiles/harmonical1D/inifile.dat",secondread=True)
grids=grid.creategrid(parameters)

#Load the files
parameters[6][0]=['from-file',parameters[10]+'/field.npy']
psi=fields.spin0(parameters, grids)
pot=np.load(parameters[10]+'/potential.npy')
energy=np.load(parameters[10]+'/energy.npy')
totalmass=np.load(parameters[10]+'/totalmass.npy')
timeline=np.load(parameters[10]+'/timeline.npy')
```

This method shoud give the ground state, the function associated is
$$\psi_{d=1}=\bigg(\frac{\omega}{\pi}\bigg)^{1/4}e^{-\frac{1}{2}\omega x^2}$$
and the energy for $d=1$ is
$$E_{d=1}=\frac{1}{2}\omega$$
for $d=2$:
$$\psi_{d=2}=\bigg(\frac{\omega}{\pi}\bigg)^{1/2}e^{-\frac{1}{2}\omega (x^2+y^2)}$$
$$E_{d=2}=\omega$$
and for $d=3$:
$$\psi_{d=3}=\bigg(\frac{\omega}{\pi}\bigg)^{3/4}e^{-\frac{1}{2}\omega (x^2+y^2+z^2)}$$
$$E_{d=3}=\frac{3}{2}\omega$$

```python
#Density plot
if parameters[0]==1:
    efe=((parameters[8][2]/np.pi)**0.25)*np.exp(-0.5*parameters[8][2]*grids[2]**2)
    thenergy=0.5*parameters[8][2]

    fig = plt.figure(figsize=[4,3.5])
    plt.xlabel("x")
    plt.ylabel(f'$\psi$')
    plt.plot(grids[2],efe, label="analytical")
    plt.plot(grids[2],np.sqrt(psi.re**2+psi.imag**2),'-.', label="numerical")
    plt.legend()
    plt.show()
    
elif parameters[0]==2:
    efe2=np.sqrt(parameters[8][2]/np.pi)*np.exp(-0.5*parameters[8][2]*(grids[2]**2+grids[4]**2))
    thenergy=parameters[8][2]

    figpot, (ax1pot, ax2pot) = plt.subplots(1, 2, figsize=(8,3.5))
    ax1pot.set_xlabel("x")
    ax2pot.set_xlabel("y")
    ax1pot.set_ylabel(f'$\psi$')
    ax2pot.set_ylabel(f'$\psi$')
    ax1pot.plot(grids[4][0],efe2[:,np.where(efe2==np.amax(efe2))[0][0]], label="analytical")   
    ax2pot.plot(grids[4][0],efe2[np.where(efe2==np.amax(efe2))[1][0],:], label="analytical")   
    ax1pot.plot(grids[4][0],np.sqrt(psi.re**2+psi.imag**2)[:,np.where(efe2==np.amax(efe2))[0][0]],'-.', label="numerical")   
    ax2pot.plot(grids[4][0],np.sqrt(psi.re**2+psi.imag**2)[np.where(efe2==np.amax(efe2))[1][0],:],'-.', label="numerical")
    ax1pot.legend()
    ax2pot.legend()

    figpot.tight_layout(pad=2.0)
    
    fig = plt.figure(figsize=(6,6))
    ax = fig.add_subplot(projection='3d')
    ax.view_init(10, -5)
    ax.scatter(grids[2], grids[4], efe2, alpha=0.9, label="analytical")
    ax.scatter(grids[2], grids[4], np.sqrt(psi.re**2+psi.imag**2), alpha=0.1, label="numerical")

    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('f')
    ax.legend()
    plt.show()
    
    fig2, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,3))
    fig2a=ax1.contourf(grids[2][:,np.where(efe2==np.amax(efe2))[1][0]], grids[4][np.where(efe2==np.amax(efe2))[0][0],:], efe2)
    fig2b=ax2.contourf(grids[2][:,np.where(efe2==np.amax(efe2))[1][0]], grids[4][np.where(efe2==np.amax(efe2))[0][0],:], np.sqrt(psi.re**2+psi.imag**2))
    plt.colorbar(fig2a, ax=ax1)
    plt.colorbar(fig2b, ax=ax2)
    plt.axis('scaled')
    plt.show()

elif parameters[0]==3:
    efe3=((parameters[8][2]/np.pi)**0.75)*np.exp(-0.5*parameters[8][2]*(grids[2]**2+grids[4]**2+grids[6]**2))
    thenergy=1.5*parameters[8][2]
    
    figpot, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12,4))
    ax1.plot(grids[6][0][0],efe3[:,np.where(efe3==np.amax(efe3))[1][0],np.where(efe3==np.amax(efe3))[2][0]], label="analytical")   
    ax2.plot(grids[6][0][0],efe3[np.where(efe3==np.amax(efe3))[0][0],:,np.where(efe3==np.amax(efe3))[2][0]], label="analytical")
    ax3.plot(grids[6][0][0],efe3[np.where(efe3==np.amax(efe3))[0][0],np.where(efe3==np.amax(efe3))[1][0],:], label="analytical")  
    ax1.plot(grids[6][0][0],np.sqrt(psi.re**2+psi.imag**2)[:,np.where(efe3==np.amax(efe3))[1][0],np.where(efe3==np.amax(efe3))[2][0]],'-.', label="numerical")   
    ax2.plot(grids[6][0][0],np.sqrt(psi.re**2+psi.imag**2)[np.where(efe3==np.amax(efe3))[0][0],:,np.where(efe3==np.amax(efe3))[2][0]],'-.', label="numerical")
    ax3.plot(grids[6][0][0],np.sqrt(psi.re**2+psi.imag**2)[np.where(efe3==np.amax(efe3))[0][0],np.where(efe3==np.amax(efe3))[1][0],:],'-.', label="numerical")

    ax1.set_xlabel('x')
    ax1.set_ylabel(f'$\psi$')
    ax1.legend()
    ax2.set_xlabel('y')
    ax2.set_ylabel(f'$\psi$')
    ax2.legend()
    ax3.set_xlabel('z')
    ax3.set_ylabel(f'$\psi$')
    ax3.legend()
    
    figpot.tight_layout(pad=2.0)
```
